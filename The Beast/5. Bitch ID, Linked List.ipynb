```python
#singly linked list with head and tail to append more efficiently

class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    
class SpecialSinglyLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None
    self.size = 0

  def print_all(self):
    current_node = self.head
    while current_node:
      if(current_node.next is None):
        print(current_node.data)
      else:
        print(current_node.data, end='->')
      current_node = current_node.next

  # Append data to the end of the linked list;
  def append(self, new_data):
    new_node = Node(new_data)
    if self.head is None:
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node
      self.tail = new_node
    self.size += 1

sll = SpecialSinglyLinkedList()
sll.append(20)
sll.append("a")
sll.append(3.45)
sll.print_all()    
```

```python
#Double Linked List

class ListNode:
  def __init__(self, data):
      self.data = data
      self.prev = None
      self.next = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None
    self.size = 0
    
  def append(self, data):
    node = ListNode(data)
    if self.head is None:
      self.head = node
      self.tail = node
    else:
      self.tail.next = node
      node.prev = self.tail
      self.tail = node
    self.size += 1
    
  def delete(self, data):
    current = self.head
    while current is not None:
      if current.data == data:
        if current == self.head:  # Deletion at the beginning of the list
          print("in head")
          self.head = current.next  # Update head
          if current == self.tail:  # Only a single element
            self.tail = None
          else:
            self.head.prev = None
        elif current == self.tail:  # Deletion at the end of the list
          current.prev.next = None
          self.tail = current.prev 
          print("in tail") # Update tail
        else:  # Delection in the middle of the list
          current.prev.next = current.next
          current.next.prev = current.prev
          print("in middle")
        self.size -= 1
        return
      current = current.next
  
  def print_all(self):
    current_node = self.head
    while current_node:
      print(current_node.data, end ="->")
      current_node = current_node.next
    print()


  def print_reverse(self):
    current_node = self.tail
    while current_node:
      print(current_node.data)
      current_node = current_node.prev

  # Search to see if the list contains data
  def search(self, data):
    current_node = self.head
    while current_node is not None:
      if current_node.data == data:
        return True  
      current_node = current_node.next
    return False
  
  # Clear the whole list
  def clear(self):
    self.head = None
    self.tail = None
  # insertBefore(self, givendata, newdata)
  def insertBefore(self, givendata, newdata):
    
    current_node = self.head
    #search the node with givendata
    while current_node is not None:
      if current_node.data == givendata: #found the node. We need to insert newdata before it
        newnode = ListNode(newdata)
        if current_node == self.head: # the givendata is in the head
           newnode.next = self.head
           self.head.prev = newnode
           self.head = newnode
           return
        else:  
          
          current_node.prev.next = newnode
          newnode.prev = current_node.prev

          current_node.prev = newnode
          newnode.next = current_node
          return
      else: 
        current_node = current_node.next
    print("Not found")

  def __iter__(self):
    current = self.head
    while current is not None:
      data = current.data
      current = current.next
      yield data
      
      
dll = DoublyLinkedList()
for color in ("yellow", "blue", "purple", "green", "black", "orange"):
    dll.append(color)
print("doubly linked list")
dll.print_all()
print(dll.search("purple"))
print(dll.search("PurPLE"))
print("doubly linked list after delete")
dll.delete("yellow")
dll.print_all()
print("Reverse: .......")
dll.print_reverse()
dll.insertBefore("yellow", "red")
dll.insertBefore("orange", "pink")
print("doubly linked list after insertbefore()")
dll.print_all()
```
