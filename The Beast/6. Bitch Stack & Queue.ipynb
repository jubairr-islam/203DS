#**Stack & Queue Functions**#

```python
class Stack:
  def __init__(self):
    self.elements = []
  
  def pop(self):
    return self.elements.pop()
  
  def peek(self):
    return self.elements[-1]

  def push(self,v):
    self.elements.append(v)
  
  def isEmpty(self):
    return len(self.elements) == 0

s = Stack()
s.push(10)
s.push(11)
s.push(12)
print(s.pop())
print(s.pop())
print(s.pop())

class Queue:
  def __init__(self):
    self.elements = []
  
  def dequeue(self):
    return self.elements.pop(0)
  
  def enqueue(self,v):
    self.elements.append(v)
  
  def isEmpty(self):
    return len(self.elements) == 0

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.isEmpty())
```

```python
from collections import deque

# stack implementation using deque (Double ended queue)
class StackDe:
    def __init__(self):
        self.data = deque()
        
    def size(self):
        return len(self.data)

    def isempty(self):
      if self.size()==0:
        return True
      else: 
        return False 
    
    def push(self, data):
        self.data.append(data)

    def pop(self):
      if self.isempty():
        print("Stack is empty")
      else:
        self.data.pop()

    def peek(self):
      if self.isempty():
        print("Stack is empty")
      else:
        return self.data[-1]

    def printStack(self):
      if self.size() == 0:
        print('Stack is empty')
      else:
        print(self.data)

s= StackDe()

s.peek()
s.printStack()
s.push('a')
s.push('b')
s.push('c')
s.push('d')
s.printStack()
```

```python
from collections import deque

# Queue implementation using deque (Double ended queue)
class QueueDe:
    def __init__(self):
        self.__data = deque()
        
    def size(self):
        return len(self.__data)

    def isempty(self):
      if self.size()==0:
        return True
      else: 
        return False 
    
    def enqueue(self, data):
        self.__data.append(data)

    def dequeue(self):
      if self.isempty():
        print("Queue is empty")
      else:
        self.__data.popleft()

    def peek(self):
      if self.isempty():
        print("Queue is empty")
      else:
        return self.__data[0]      

    def printQueue(self):
       if self.isempty():
        print("Queue is empty")
       else:
        print(self.__data)   


q = QueueDe()
q.peek()
q.dequeue()
q.enqueue("Aquib")
q.enqueue("Arif")
q.enqueue("Sajib")
q.enqueue("Nafis")
#q.peek()
q.dequeue()
q.printQueue()

# print(q._QueueDe__data)
```
